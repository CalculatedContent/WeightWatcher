```python
x, mp = RMT_Util.marchenko_pastur_pdf(0.1, 5, Q=1, sigma=1)
plt.plot(x,mp, linewidth=1, color = 'r', label="$Q=1;\sigma^{2}=1$")

x, mp = RMT_Util.marchenko_pastur_pdf(0.1, 5, Q=4, sigma=1)
plt.plot(x,mp, linewidth=1, color = 'b', label="$Q=4;\sigma^{2}=1$")

x, mp = RMT_Util.marchenko_pastur_pdf(0.1, 5, Q=10, sigma=1)
plt.plot(x,mp, linewidth=1, color = 'g', label="$Q=10;\sigma^{2}=1$")
         
plt.title(r"Marchenko Pastur distributions $\rho(\lambda)$")
plt.xlabel(r"$\lambda\in[\lambda^{-},\lambda^{+}] $")
plt.ylabel(r"$\rho(\lambda)$")
plt.legend()
plt.tight_layout()

filename="img/mp-Q.png"
plt.savefig(filename)

print("saved figure {}".format(filename))
```
